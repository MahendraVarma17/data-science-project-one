```python
import spacy
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
import string

# Load spaCy model and stopwords
nlp = spacy.load('en_core_web_sm')
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    doc = nlp(text)
    tokens = [token.lemma_ for token in doc if token.text not in stop_words and token.text not in string.punctuation]
    return ' '.join(tokens)
```

```python
sample_text = "Neural networks are computer systems modeled after the human brain. They consist of layers of nodes that work together to recognize patterns. This technology powers applications like image recognition and natural language processing"
print(preprocess_text(sample_text))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

def extract_keywords(text, n=5):
    vectorizer = TfidfVectorizer(max_features=10)
    tfidf_matrix = vectorizer.fit_transform([text])
    keywords = vectorizer.get_feature_names_out()
    return keywords[:n]

# Example usage
preprocessed_text = preprocess_text(sample_text)
print(extract_keywords(preprocessed_text))
```

```python
from transformers import T5ForConditionalGeneration, T5Tokenizer

# Load T5 model and tokenizer
model = T5ForConditionalGeneration.from_pretrained('t5-small')
tokenizer = T5Tokenizer.from_pretrained('t5-small')

def generate_title(text):
    input_text = "summarize: " + text
    inputs = tokenizer.encode(input_text, return_tensors="pt", max_length=512, truncation=True)
    outputs = model.generate(inputs, max_length=10, num_beams=4, early_stopping=True)
    title = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return title

# Example usage
print(generate_title(preprocessed_text))
```
