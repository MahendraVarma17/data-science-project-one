```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import read_csv
path="/content/drive/MyDrive/energy_dataset (1).csv"
electricity=pd.read_csv(path)
print(electricity)
print(electricity.shape)
print(electricity.head(5))
print(electricity.tail(4))
print(electricity.info())
print(electricity.describe())
print('column date')
print(electricity.columns)
print("++++++++++++++++++++++++++++")
print(electricity[electricity['State'].str.startswith('M')])
print("++++++++++++++++++++++++++++")
print(electricity[electricity['State'].str.endswith('a')])
print("++++++++++++++++++++++++++++")
rcb=electricity[electricity['State']=='Andhra']
print(rcb)
print("+++++++++++++++++++++++++++++++")

rcb=electricity[electricity['State']=='Manipur']
print(rcb)
print(electricity.sort_values(by='State'))
print(np.max(electricity['Consumption']))
print(np.min(electricity['Consumption']))
print(np.mean(electricity['Consumption']))
state_data=electricity[electricity['State']=='Andhra']
print(state_data)
state_data2=electricity[electricity['State']=='Karnataka']
state_data3=electricity[electricity['State']=='Telangana']
state_data4=electricity[electricity['State']=='Tripra']
fig,axes=plt.subplots(2)
axes[0].plot(state_data['Date'],state_data['Consumption'])
axes[1].plot(state_data2['Date'],state_data['Consumption'])
#bar chart
fig,axes1=plt.subplots(2)
axes1[0].bar(state_data3['Date'],state_data3['Consumption'])
axes1[1].bar(state_data4['Date'],state_data4['Consumption'])
#pie chart
fig,axes2=plt.subplots(2)
state1='UP'
state2='Kerala'
state3='Mizoram'
state4='Delhi'
labels=[state1,state2]
state1_total=electricity[electricity['State']== state1]
state2_total=electricity[electricity['State']== state2]
energy=[state1_total,state2_total]
axes2[0].pie(energy,labels=labels,autopct='%1.1f%%')
```
